## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
## 1. My Lane as an ML Task (Type)

**Task Type:** Scoring (Ranking)

I selected **Scoring (Ranking)** because the objective is to assign each content page a priority score and rank the pages based on their need for review. Instead of simply predicting whether a page should be refreshed or not, the model helps prioritize pages so that content teams can review the highest-impact opportunities first. This approach supports better decision-making by producing an ordered list of recommendations based on observable search and engagement signals.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
## 2. Target or Proxy

**Target:** Predict whether a content page should be prioritized for review based on its likelihood of declining performance.

For the starter dataset, the target is a **proxy label** derived from the `trend_direction` column. A page is labelled as declining when `trend_direction == "down"`. This is a **defined rule**, not a future observed outcome. It is used as a simple proxy in the starter dataset to demonstrate the complete machine learning workflow. In a stronger capstone project, the target would be based on a future observed outcome, such as whether a page declines or recovers during a later time window.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
## 3. Success Metric

**Success Metric:** Precision@50

I will use **Precision@50** as the primary evaluation metric because the goal is to rank content pages and recommend the most important ones for review. Precision@50 measures how many of the top 50 recommended pages are actually good candidates for content refresh. A higher Precision@50 indicates that the model is effectively prioritizing the pages that need attention, making it a suitable metric for this decision-support task.
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
## 5. Why ML Beats a Fixed Rule Here

A fixed rule, such as recommending pages with impressions below a certain threshold, cannot capture the complex relationships between multiple search and engagement signals. Content performance depends on several factors, including impressions, clicks, sessions, average position, click-through rate, and trend direction, which interact in different ways. Machine learning can learn these patterns from historical data and assign a priority score that better identifies which pages should be reviewed first. This results in more accurate and consistent recommendations than a simple if-statement.
```
